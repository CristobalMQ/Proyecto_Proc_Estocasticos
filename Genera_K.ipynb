```python
import pandas as pd
import numpy as np
```

```python
lista_pasos = [10, 100, 1_000, 10_000, 100_000, 1_000_000]

files = [
"Resultados_MonteCarlo\\10_pasos.csv",
"Resultados_MonteCarlo\\100_pasos.csv",
"Resultados_MonteCarlo\\1000_pasos.csv",
"Resultados_MonteCarlo\\10000_pasos.csv",
"Resultados_MonteCarlo\\100000_pasos.csv",
"Resultados_MonteCarlo\\1000000_pasos.csv",
]

names=["i", "count", "freq"]

df = pd.DataFrame()
df["i"] = np.array([j for j in range(1,21)])

for idx,file_name in enumerate(files):
    df_csv = pd.read_csv(file_name, sep=";", names=names, header=None)
    df_csv = df_csv[["i","freq"]]
    df = df.merge(df_csv, on="i", how="right", suffixes=(f"_x{lista_pasos[idx]}", f"_y{lista_pasos[idx]}"))
```

```python
df = df.rename(columns={
    "freq_x100"    : 10,
    "freq_y100"    : 100,
    "freq_x10000"  : 1_000,
    "freq_y10000"  : 10_000,
    "freq_x1000000": 100_000,
    "freq_y1000000": 1_000_000,
})

df = df.fillna(0)
df
```

```python
df.to_excel("res.xlsx")
```
